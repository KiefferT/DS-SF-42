```python
# We've got our training and test set, time to define some functions
blah = [1., 4., 7., 9.]
blah2 = [2., 4., 3., 7.]

def CalcDist(row1, row2):    
    diffList = []
    for i in range(len(row1)):
        dist_sq = (row1[i] - row2[i]) ** 2
        diffList.append(dist_sq)
    return np.sqrt(sum(diffList))
```

```python
CalcDist(blah, blah2)
```

```python
def KNearest(test_df, train_df, target, K):
    pred_df = test_df
    
    # drop off target(y) from dfs
    test_no_y = test_df.drop([target], axis = 1)
    train_no_y = train_df.drop([target], axis = 1)
    
    predictions = []
    for x in range(len(test_df)):
        test_row = test_no_y.iloc[x].values.tolist()                  
        
        distances = []
        for y in range(len(train_df)):
            train_row = train_no_y.iloc[y].values.tolist() 
            distances.append([y, CalcDist(test_row, train_row)])  # Adds Row_number (training df), distance
        
        distances = pd.DataFrame(distances, columns = ['train_row', 'dist']) # Set distance info to df
        d2 = distances.sort_values(by = 'dist')[0:K] # Sort distance df and get closest 5 K rownums
        
        closest_trainers = train_df.iloc[d2['train_row']][target] # Get class from K nearest
                       
        prediction =  closest_trainers.value_counts().idxmax() # Get highest 'voted'
        
        predictions.append(prediction)
    
    pred_df['pred'] = predictions
    pred_df['correct'] = pred_df[target] == pred_df['pred']
    
    print "Running Kief's Nearest Neighbors"
    print "Predicting classes for", test_df.shape[0], "cases"
    print "against a training set of", train_df.shape[0], "training observations"
    print "----------------------------------------------------------------------"
    print "Correct Classifications:", sum(pred_df['correct'])
    print "Accuracy:", round((float(sum(pred_df['correct'])) / pred_df.shape[0]) * 100.0, 2), "%"
    return pred_df
```
